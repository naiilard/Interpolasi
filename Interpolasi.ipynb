```python
# Interpolation

import matplotlib.pyplot as plt
import numpy as np
from math import pi

f = lambda x: np.sin(x)
x_min, x_max = 0, pi
npt0 = 100 # fungsi ideal
npt1 = 4 # titik yang ada

x0 = np.linspace(x_min, x_max, npt0)
x1 = np.linspace(x_min, x_max, npt1)

plt.plot(x0, f(x0), 'b')
plt.plot(x1, f(x1), 'ro')

# Linear Interpolation
x2 = []
y2 = []
npt2 = 10

for i in range(npt1-1):
    slope = (f(x1[i+1]) - f(x1[i])) / (x1[i+1] - x1[i])
    delta = (x1[i+1] - x1[i]) / npt2
    for j in range(npt2+1):
        x2.append(x1[i] + j*delta)
        y2.append(f(x1[i]) + j*slope*delta)

plt.title('Linear Interpolation')
plt.plot(x2, y2, '*', color = 'red')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import CubicSpline

x = [-5, -1, 0, 2]
y = [-2, 6, 1, 3]

f = CubicSpline(x, y)
x_new = np.linspace(-5, 2, 1000)
y_new = f(x_new)

arr = np.array(x_new)
indx = np.argmax(y_new)
print('Nilai maksimum adalah', arr[indx], y_new.max())
print('Nilai minimum adalah', arr[indx], y_new.min())

plt.title('Cubic Spline')
plt.plot(x, y, 'ro')
plt.plot(x_new, y_new, color = 'blue')
plt.show()
```

```python
# Tugas
# 1 Lakukan interpolasi cubic spline untuk data spline-data, tentukan titik ekstremnya

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.interpolate import CubicSpline

# Menginput nilai x dan y dari file excel
df = pd.read_excel("/content/spline-data.xlsx")
x = df['x'].values
y = df['y'].values

# Melakukan interpolasi cubic spline
f = CubicSpline(x, y)
x_new = np.linspace(x.min(), x.max(), 1000)
y_new = f(x_new)

# Menentukan nilai ekstrem
ekstrem = []
for i in range(len(x_new)):
    if y_new[i] == y_new.max() or y_new[i] == y_new.min():
        ekstrem.append(x_new[i])
        ekstrem.append(y_new[i])
        print("\nTitik ekstrem yang diperoleh adalah")
        print(x_new[i], y_new[i])

# Membuat grafik
plt.title('Cubic Spline')
plt.plot(x, y, 'ro')
plt.plot(x_new, y_new, color = 'blue')
plt.plot(ekstrem[::2], ekstrem[1::2], 'go')
plt.legend(['Data', 'Interpolasi', 'Ekstrem'])
plt.show()
```

```python
# Latihan
# 2 Fit the data using Cubic Spline Interpolation and find minimum point

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.interpolate import CubicSpline

# Menginput data x dan y
x = [0, 1, 2, 3, 4]
y = [2, 1, 3, 5, 1]

# Melakukan interpolasi cubic spline
f = CubicSpline(x, y)
x_new = np.linspace(0, 4, 1000)
y_new = f(x_new)

# Menentukan nilai minimum
arr = np.array(x_new)
indx = np.argmin(y_new)
print('Nilai minimum adalah', arr[indx], y_new.min())

# Membuat grafik
plt.title('Cubic Spline')
plt.plot(x, y, 'ro')
plt.plot(x_new, y_new, color = 'blue')
plt.legend(['Data', 'Interpolasi'])
plt.show()
```
